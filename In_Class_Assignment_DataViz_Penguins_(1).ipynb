## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',      # color by species
    palette='Set2',     # nice colors
    s=100               # size of points
)
plt.title('Scatter plot of Bill Length vs Bill Depth by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()
```

**Your interpretation (2–3 sentences):**
There is a clear separation of species in terms of bill dimensions.

Adelie penguins tend to have shorter bills and deeper bills, while Gentoo penguins have longer bills with moderate depth.

Chinstrap penguins fall between the other two species but have relatively consistent bill depths.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

plt.figure(figsize=(8,5))
sns.histplot(penguins['bill_length_mm'], bins=10, kde=False, color='skyblue')
plt.title('Histogram of Bill Length (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()
```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(8,5))
sns.histplot(penguins['bill_length_mm'], bins=30, kde=False, color='salmon')
plt.title('Histogram of Bill Length (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()

```

**Your comparison (1–2 sentences):**

Write your observatioWith 10 bins, the histogram shows a smoother overall distribution, making it easier to see general trends.

With 30 bins, smaller variations and peaks in the data become visible, but the histogram can appear noisier.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

plt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    palette='pastel'
)
plt.title('Body Mass Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

**Your summary (1–2 sentences):**

Gentoo penguins are the heaviest on average, while Adelie penguins are the lightest.

Chinstrap penguins have a body mass distribution that falls between the other two species, with moderate variation.
